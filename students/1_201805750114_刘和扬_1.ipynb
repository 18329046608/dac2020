### 问题1：用``urllib.requests``库发起一次``get``请求，输出``response``文本信息。

```python
#建立类
class Class1():
    def __init__(self,url):
        self.url=url       #设立属性
        
#用urllib.requests库发起一次get请求，输出response文本信息        
    def a(self,timeout):
        import urllib.request
#data=bytes(urllib.parse.urlencode({'dghgdh':'hdjd'}),encoding='utf8')  
#初始化满足utf8接口的bytearray并转化为二进制数据
#发送get请求不能设置data参数如果设置了data则为post请求，但即使url最后加/get好像添加data参数并不会报错
        response=urllib.request.urlopen(self.url+'get',timeout=timeout)  
    #设置了超时释放连接为0.5s可以省去，发起请求且访问 在self.url后面加get与不加get都可以输出而不报错但输出结果不一样
        return response
    
# 用Requests库伪装成火狐浏览器发起一次post请求，输出response文本信息   
    def b(self,User,host,data):
        import requests
        headers={'User- Agent':User,"host":host}
#伪装成火狐浏览器，host 是指发送请求地址
        postxingxi=requests.post(self.url,headers=headers,data=data) #发送get请求还可添加params参数注意data参数post才带data参数
        return postxingxi
    
#  用Requests库中的session对象发出get请求，设置cookies并获取，输出获取的cookies内容  
    def c(self,cookies):
        import requests
        requests.Session().get(self.url+'cookies/set/sessioncookie/'+cookies)
#发出请求建立cookie为39883839
        cook=requests.Session().get(self.url+"cookies")
#发出另一请求获取cookie
        return cook

```

```python
# 这里编写代码
getxingxi=Class1('http://www.baidu.com/').a(0.5)
#调用Class类中的方法a
#设立初始网站这个网站设置要注意如果这个网站被限制了浏览器会报HTTP Error 403: Forbidden错比如http://www.chinadaily.com.cn/这个网站
print(getxingxi.read().decode('utf-8'))#当不设置decode为utf-8时输出很乱
```

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
import requests,re
count=0#控制层数
k=2#控制爬取层数
r=re.compile(r'href=[\'"]?(http[^\'">]+)')#用正则表达式得到Pattern对象,,第一个r是表示字符串不会转义r也可以用u
stack=[['http://httpbin.org/']]#设立种子网站存在多层列表里，每层放相应层url
#存储所要爬取的k+1层的url
while count<k:
    try:
        url=stack[count].pop(0)#弹出每层第一个url
        print(url)#输出用来生成吓一层的url
        html=requests.get(url).text#用正则表达式过滤url的内容
        new_urls=r.findall(html)#探索下一层
        stack.append(new_urls)#保存新探索的url
        count+=1#计数器加1
    except Exception as e:
#捕获除与程序退出sys.exit()相关之外的所有异常。                
        print(url)
        print(e)
#按深度遍历输出网站
while count!=0:
    if len(stack[count])!=0:
        url=stack[count].pop(0)#弹出每层第一个url
        print(url)#输出url
        html=requests.get(url).text#用正则表达式过滤url的内容
    else:
        count-=1
```

输出正确不过只输出url未输出被访问的内容，且输出结果较少这个可以通过加大爬取层数k改变

这里对结果进行分析。

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
#去重算法
import requests,re
count=3#控制层数为count层
r=re.compile(r'href=[\'"]?(http[^\'">]+)')#用正则表达式得到Pattern对象,,第一个r是表示字符串不会转义r也可以用u
stack=['http://httpbin.org/']#设立种子网站存在列表里
used=set()#设立集合用集合是因为它可哈希，不重复
storage=[]#建立列表
while len(stack)>0 and count>0:
    try:
        url=stack.pop(0)#弹出列表第一个url
        html=requests.get(url).text#用正则表达式过滤url的内容
        new_urls=r.findall(html)#探索下一层
        used.add(url)#将已抓取url存入集合
        storage.append(url)#保存弹出的url
        #去重我尝试了一下直接把storage改变成集合也可以达到在最后去重的效果不过直接转化成集合会让stack花费更大空间和不去重算法一样
        for new_url in new_urls:
            if new_url not in used and new_url not in stack:
                stack.append(new_url)#把新探索的url加入stack
        count-=1#计数器减1
    except Exception as e:
#捕获除与程序退出sys.exit()相关之外的所有异常。                
        print(url)
        print(e)
storage.extend(stack) #由于是广度优先所以只要按照stack的顺序输出
for i in range(len(storage)):
    print(storage[i])#输出url
print(len(storage))
```

### (选)将抓取网页的去重深度遍历算法封装成对象（类），并测试。 

```python
# 这里编写代码
class shengdubianli():
    def __init__(self,y):
        self.y=y      #设立属性
    def quchong(self,k):
        import requests,re
        count=0#控制层数
        k=k#控制爬取层数
        used=set()#用来存放访问过的url
        r=re.compile(r'href=[\'"]?(http[^\'">]+)')#用正则表达式得到Pattern对象,,第一个r是表示字符串不会转义r也可以用u
        seed=self.y
        stack=[[seed]]#设立种子网站存在多层列表里，每层放相应层url
#存储所要爬取的k+1层的url
        while count<k:
            try:
                url=stack[count].pop(0)#弹出每层第一个url
                print(url)#输出用来生成吓一层的url
                used.add(url)#把访问过的url放入
                html=requests.get(url).text#用正则表达式过滤url的内容
                new_urls=r.findall(html)#探索下一层
            #把已经访问过或存放在stack中的url从new_urls中删去
                for new_url in set(new_urls):
                    if new_url in used or new_url in stack:
                        used.remove(url)
                stack.append(new_urls)#保存新探索的不重复的url
                count+=1#计数器加1
            except Exception as e:
#捕获除与程序退出sys.exit()相关之外的所有异常。                
                print(url)
                print(e)
#按深度遍历输出网站
        while count!=0:
            if len(stack[count])!=0:
                url=stack[count].pop(0)#弹出每层第一个url
                print(url)#输出url
                html=requests.get(url).text#用正则表达式过滤url的内容
            else:
                count-=1
```
