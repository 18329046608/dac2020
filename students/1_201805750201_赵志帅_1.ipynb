### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
import requests, re
from collections import Counter
count = 15
seed = "https://www.163.com/" 
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')

stack = [seed]
storage = {}
while len(stack) > 0 and count > 0:
    try:
        url = stack.pop(-1)
        html = requests.get(url).text
        new_urls = r.findall(html)
        stack.extend(new_urls)
        storage[url] = len(new_urls)
        count -= 1
    except Exception as e:
        print(url)
        print(e)

url_count = Counter(stack)
for url,count in url_count.most_common(15):
    print(url,count)
```
