### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
from collections import Counter
import requests, re
count = 15
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')  #正则表达式
seed = 'https://www.baidu.com/' 
stack = [seed]                                 #种子网站入栈
storage = {}
while len(stack) > 0 and count > 0:
    try:                                       #异常处理
        url = stack.pop(-1)                    #用栈处理时pop()中值为-1
        html = requests.get(url).text
        new_urls = r.findall(html)
        stack.extend(new_urls)
        storage[url] = len(new_urls)
        count -= 1
    except Exception as e:
        print(url)
        print(e)
        
url_count = Counter(stack)                     #进行计数
for url,count in url_count.most_common(15):
    print(url,count)
```

这里对结果进行分析。所输出结果即为对百度进行深度遍历得到的网址及其数目。

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
#不去重
from collections import Counter                     #引入计数
import requests
import re
count = 10
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')
seed = 'https://www.baidu.com/' 
queue = [seed]                                      #此时用队列存放种子网站
storage = {}
while len(queue) > 0 and count > 0:
    try:
        url = queue.pop(0)                          #用队列处理时pop()中值为0
        html = requests.get(url).text
        new_urls = r.findall(html)
        queue.extend(new_urls)
        storage[url] = len(new_urls)
        count -= 1
    except Exception as e:
        print(url)
        print(e)
url_count = Counter(queue)                           #计数
for url,count in url_count.most_common(15):
    print(url,count)
```

```python
#去重
from collections import Counter
import requests
import re
count = 10
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')
seed = 'https://www.baidu.com/' 
queue = [seed]
used=set()                                                      #创建used集合，用于存放未抓取的不重复的url
storage = {}
while len(queue) > 0 and count > 0:
    try:
        url = queue.pop(0)
        html = requests.get(url).text
        storage[url]=html
        used.add(url)                                           #将已抓取的url存入used集合中
        new_urls = r.findall(html)
        for new_url in new_urls:                                #将未抓取的url添加到stack中
            if new_url not in used and new_url not in queue:    #判断捕获的url是否在队列中，不在则存入
                queue.append(new_url)
        count -= 1
    except Exception as e:
        print(url)
        print(e)
url_count = Counter(queue)                                       #用于计数
for url,count in url_count.most_common(15):
    print(url,count)
```

这里对结果进行分析。很明显，在用广度遍历不去重时，所得到的数目有重复，但去重之后数目均为一。
