### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
import requests
import re
count = 0
depth = int(input('Please input the maxmum depth:'))   #输入最大深度                                 
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:73.0) Gecko/20100101 Firefox/73.0'}#伪装成火狐浏览器
#url = 'http://wjw.jiangsu.gov.cn/'
url = 'https://tieba.baidu.com/index.html?traceid=' #种子网站
urlDict = [url]  
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')
old_urls = []
while(count <= depth and len(urlDict) > 0):
    try:
        url_another = urlDict.pop(-1)#从列表中的最后一个输出
        old_urls.append(url_another)
        print(url_another)
        html = requests.post(url_another,headers = headers).text
        urls = r.findall(html)
        urls.reverse()#将列表倒置，以实现深度优先遍历
        for url in urls:
            if (url not in old_urls):
                urlDict.append(url)
        count += 1
    except Exception as e:
        pass
```

结果分析：深度遍历出来的网站相较于广度遍历的网站少。
