### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
##先用用浏览器打开 https://httpbin.org/headers 网站，就会出现所要伪装的浏览器IDEheaders
import requests
url='http://www.nhc.gov.cn/'
##通过设置header,将爬虫伪装成浏览器
send_headers={
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8", 
    "Accept-Encoding": "gzip, deflate, br", 
    "Accept-Language": "zh-CN,zh;q=0.8,zh-TW;q=0.7,zh-HK;q=0.5,en-US;q=0.3,en;q=0.2", 
    "Host": "httpbin.org", 
    "Upgrade-Insecure-Requests": "1", 
    "User-Agent": "Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:73.0) Gecko/20100101 Firefox/73.0", 
    "X-Amzn-Trace-Id": "Root=1-5e635aff-66dfefa0b923e90c8d7e788c"
}
#发起一次post请求
r=requests.post(url,headers=send_headers)
r.text
```

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
#不去重深度遍历算法
import requests,re
count=10
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
seed='http://httpbin.org/'
stack=[seed]
storage={}
while len(stack)>0 and count>0:
    try:
        url=stack.pop(-1)
        html=requests.get(url).text
        new_urls=r.findall(html)
        stack.extend(new_urls)
        print(url)
        storage[url]=len(new_urls)
        count-=1
    except Exception as e:
        print(url)
        print(e)
```

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
#去重的广度遍历算法
import requests,re
count=3
recount=0
allcount=1
r=re.compile(r'href=[\'"]?(http[^\'">]+)')#根据包含的正则表达式的字符串创建模式对象
seed='http://hao.360.com/'#以'http://hao.360.com/'作为种子网站
queue=[seed]#运用队列保存种子网站
used=set()#存放已爬取的url
storage={}
while len(queue)>0 and count>0:
    try:
        url=queue.pop(0)
        html=requests.get(url).text
        storage[url]=html
        used.add(url)
        new_urls=r.findall(html)
        for new_url in new_urls:
            allcount+=1
            if new_url not in used and new_url not in queue:
                queue.append(new_url)
                print(new_url)
            else:
                print("重复："+new_url)
                recount+=1
        count-=1
    except Exception as e:
        print(url)
        print(e)
```
