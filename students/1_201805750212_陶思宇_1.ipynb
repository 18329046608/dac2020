### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
# 这里编写代码
```

```python
from urllib import request,parse
url="https://www.amazon.cn/s?k=kindle&__mk_zh_CN=%E4%BA%9A%E9%A9%AC%E9%80%8A%E7%BD%91%E7%AB%99&ref=nb_sb_noss_2"
#爬取亚马逊网站，考虑到其具有反爬机制，所以伪装成火狐浏览器进行
headers={
    "User-Agent":'Mozilla/4.0(compatiable;MSIE 5.5;Windows NT)',
    "host":"www.amazon.cn"
}
dict={
    "name":"kindle"
}
data=bytes(parse.urlencode(dict),encoding="utf8")
req=request.Request(url=url,data=data,headers=headers,method="POST")
response=request.urlopen(req)
print(response.read().decode("utf-8"))#decode是进行解码，解码时需要注意utf-8和gbk区别
```

这里对结果进行分析。

**结果：由于许多网站具有反爬机制，所以伪装成火狐浏览器。在这里，想爬取亚马逊网站中kindle的相关信息，首先是爬取成功的，返回了html文档，target指向目标kindle，后有与它相关的颜色，在库情况，电子书，保护套，评分等等一系列信息。但是也在返回结果中有一大堆没用或暂时不需要不是目标相关的信息，这一点暂时不知道如何处理，需要进一步清洗优化。**

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
```

```python
import requests,re
recount=0
allcount=1
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
#采用正则表达式
seed='https://www.xtu.edu.cn/'
queue=[seed]
storage={}
while len(queue)>0:
    try:
        url=queue.pop(0)
        html=requests.get(url,timeout=3).text#获取网页源码内容
        storage[url]=html
        new_urls = r.findall(html)#获取源码中的链接
        for new_url in new_urls:
                allcount+=1
                if new_url not in queue :#因为不去重，则不用再考虑必须未使用才可添加，只要不在队列中，都视为新url
                        queue.append(new_url)
                else:
                    print(new_url)
                    recount+=1
       
    
    except Exception as e:
            print(url)
            print(e)
```

这里对结果进行分析。

**结果:因为采用不去重的深度优先抓取网页，爬取网页为湘大官网，因为是深度优先，所以不断返回相关网址。和去重算法相比，会繁琐，不用的信息很多。**

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
```

**去重:**

```python
import requests,re
count=3
recount=0
allcount=1
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
#采用正则表达式
seed='http://httpbin.org/'
used=set()
#利用集合唯一性，从而达到去重目的
queue=[seed]
storage={}
while len(queue)>0 and count >0:
    try:
        url=queue.pop(0)
        html=requests.get(url,timeout=5).text#获取网页源码内容
        storage[url]=html
        used.add(url)
        new_urls = r.findall(html)#获取源码中的链接
        for new_url in new_urls:
                allcount+=1
                if new_url not in queue and new_url not in used:#去重！
                    queue.append(new_url)
                else:
                    print("重复"+new_url)
                    #输出重复的url
                    recount+=1
        count-=1
    
    except Exception as e:
            print(url)
            print(e)
```

**不去重：**

```python
import requests,re
count=3
recount=0
allcount=1
r=re.compile(r'href=[\'"]?(http[^\'">]+)')
#采用正则表达式
seed='http://httpbin.org/'
used=list()
#利用集合唯一性，从而达到去重目的
queue=[seed]
storage={}
while len(queue)>0 and count >0:
    try:
        url=queue.pop(0)
        html=requests.get(url,timeout=5).text#获取网页源码内容
        storage[url]=html
        used.append(url)
        new_urls = r.findall(html)#获取源码中的链接
        for new_url in new_urls:
                allcount+=1
                if new_url not in queue :#因为不去重，则不用再考虑必须未使用才可添加，只要不在队列中，都视为新url
                    queue.append(new_url)
                if new_url in queue:
                    print(new_url)
                    recount+=1
        count-=1
    
    except Exception as e:
            print(url)
            print(e)
```

这里对结果进行分析。

**结果:对比去重，不去重可以明显发现去重的优越性。在去重算法中，输出的url是整个有效url中重复的链接，去重主要依赖set（）的唯一性和if关键语句达到，else将重复url输出。在不去重算法中，因为不去重，所以不再考虑used，只要没有出现在queue中都视为新的url，并把队列中的url输出。从输出结果和去重得到的重复url进行对比，结果之间也相互验证了。在不去重结果中，多次出现https://education.github.com/，https://desktop.github.com/等等。**
