### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


````python
import requests,re
deeping={}#存储爬取到的url以及其所处的深度
def get_all_url(a):
    """传入一个url，提取并输出该网页包含的所有url
    """
    html=requests.get(a).text
    condition=re.compile(r'href=[\'"]?(http[^\'" >]+)')
    url_in_putin=condition.findall(html)
    return url_in_putin
def get_urls_deep(url,deepth=3):
    """
    深度优先爬取url的递归函数
    """
    if(deeping[url]>=deepth):#深度判定
        return
    urls=get_all_url(url)
    for next_url in urls:
        #if next_url not in deeping:
        #去重判定
            deeping[next_url]=deeping[url]+1#下层的url深度加一
            get_urls_deep(next_url)
seed='http://mnks.jxedt.com/'
deeping[seed]=0
get_urls_deep(seed,deepth=3)
print(list(deeping.keys()))
````

````python
#去重
import requests,re
pre=[]#存储所有待爬取的url
basement=[]#存储所有爬取过的url
used_urls=[]#记录已经爬取过的url
url='http://mnks.jxedt.com/'
def get_all_url(a):
    """传入一个url，提取并输出该网页包含的所有url
    """
    html=requests.get(a).text
    condition=re.compile(r'href=[\'"]?(http[^\'" >]+)')
    url_in_putin=condition.findall(html)
    return url_in_putin
basement=basement+(get_all_url('http://mnks.jxedt.com/'))#将get_all_url函数爬出的u所有url合并到列表basement中
pre=pre+(get_all_url('http://mnks.jxedt.com/'))#将get_all_url函数爬出的u所有url合并到列表pre中
new_url=list.pop(pre,0)#将pre中的第一个url赋值给new_url
for i in range(1):#因为选用的seed包含63个url，爬到下一层就包含7730个url，继续深入电脑吃不住了又不去重，所以深度设置为2
    for new_url in pre:
        if new_url not in used_urls:#重复url的判定
            next_floor_urls=get_all_url(new_url)
            basement=basement+next_floor_urls
            pre=pre+next_floor_urls
            used_url=list.pop(pre,0)
            used_urls.append(used_url)#讲已经爬取过的url加入used_urls
print(len(basement))
````

## 这里对结果进行分析:
+ 这样的去重是最蠢的直接对比，没有用到hash压缩之类的技术，显然不好实际运用
+ 对比有无去重的爬取url数目发现去重还是有一些用的，但是加了这么一段去重的判定，跑完这段程序的时间并没有久多少
