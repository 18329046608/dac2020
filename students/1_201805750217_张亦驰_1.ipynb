### 问题1：用``urllib.requests``库发起一次``get``请求，输出``response``文本信息。

```python
# 这里编写代码
import urllib.request as urs      #导入urllib.request库，重命名为urs
url = 'http://www.cninfo.com.cn/new/index'   #url赋值
headers = {
    'User-Agent':'Mozilla/4.0(compatible; MSIE 5.5; Windows NT)', 'hose':'httpbin.org'
}#报文头为火狐浏览器
req = urs.Request(url=url, headers=headers, method='GET')     #构造Request实例，以GET请求方式
response = urs.urlopen(req)   #发送请求
print(response.read().decode('utf-8'))    #打印返回来的网页信息
```

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
import requests,re    #导入requests库，re库（正则表达式）
count = 10    #记数器设置为6
r = re.compile(r'href=[\''']?(http[^\'''>]+)')    #调用re库中的compile函数，其将正则表达式创建为模式对象
headers = {
    'User-Agent':'Mozilla/4.0(compatible; MSIE 5.5; Windows NT)', 'hose':'httpbin.org'
}       #伪装成火狐浏览器
seed = 'http://www.baidu.com/'     #设置种子网站
queue = [seed]         #将种子网站放入一个队列中
while len(queue) > 0 and count > 0:    #这是一个while循环。当队列为空时且记数器为0时结束
    try:
        url = queue.pop(-1)    #使用pop方法弹出队列的最后一个元素
        html = requests.get(url,params=headers).text    #使用requests库中的get函数发送请求，并将返回的请求结果字符串化
        new_urls = r.findall(html)    #使用re库中的findall函数，将在字符串中找到的正则表达式所匹配的所有字串并返回一个列表
        queue.extend(new_urls)    #将新返回的列表加入到原始列表中
        count = count - 1    #计数器减一
    except Exception as e:    #如果try部分出错了，就执行except部分
        print(e)
    else:    #如果try部分正确就执行else部分
        print(url)
```

* 可以明显的感知这个深度遍历输出结果的顺序与下方广度遍历输出结果的顺序不同
* 深度遍历总是出队列的最后一个，广度遍历则相反

这里对结果进行分析。

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
# 不去重的广度遍历算法
import requests,re    #导入requests库，re库（正则表达式）
count = 10    #记数器设置为6
r = re.compile(r'href=[\''']?(http[^\'''>]+)')    #调用re库中的compile函数，其将正则表达式创建为模式对象
headers = {
    'User-Agent':'Mozilla/4.0(compatible; MSIE 5.5; Windows NT)', 'hose':'httpbin.org'
}       #伪装成火狐浏览器
seed = 'http://www.baidu.com/'     #设置种子网站
queue = [seed]         #将种子网站放入一个队列中
while len(queue) > 0 and count > 0:    #这是一个while循环。当队列为空时且记数器为0时结束
    try:
        url = queue.pop(0)    #使用pop方法弹出队列的第一个元素
        html = requests.get(url,params=headers).text    #使用requests库中的get函数发送请求，并将返回的请求结果字符串化
        new_urls = r.findall(html)    #使用re库中的findall函数，将在字符串中找到的正则表达式所匹配的所有字串并返回一个列表
        queue.extend(new_urls)    #将新返回的列表加入到原始列表中
        count = count - 1    #计数器减一
    except Exception as e:    #如果try部分出错了，就执行except部分
        print(e)
    else:    #如果try部分正确就执行else部分
        print(url)
```

* 结果中有六个网页是可以爬取的，其他网页受到了反爬的制裁，识破了火狐浏览器的伪装。
* 没有出现10个网页就停止的原因，我认为是因为百度网页的拓扑结构所致，即在整个广度遍历中，由于某些网页的反爬干扰而限制了爬取的广度。
* 再者就是因为因为重复爬取形成了闭环

```python
# 这里编写代码
# 不重的广度遍历算法
import requests,re    #导入requests库，re库（正则表达式）
count = 10    #记数器设置为6
r = re.compile(r'href=[\''']?(http[^\'''>]+)')    #调用re库中的compile函数，其将正则表达式创建为模式对象
headers = {
    'User-Agent':'Mozilla/4.0(compatible; MSIE 5.5; Windows NT)', 'hose':'httpbin.org'
}       #伪装成火狐浏览器
seed = 'http://www.baidu.com/'     #设置种子网站
queue = [seed]         #将种子网站放入一个队列中
storage = set()    #构造一个集合对象
while len(queue) > 0 and count > 0:    #这是一个while循环。当队列为空时且记数器为0时结束
    try:
        url = queue.pop(0)    #使用pop方法弹出队列的第一个元素
        html = requests.get(url,params=headers).text    #使用requests库中的get函数发送请求，并将返回的请求结果字符串化
        storage.add(url)   #将爬取的元素加入到集合storage中
        new_urls = r.findall(html)    #使用re库中的findall函数，将在字符串中找到的正则表达式所匹配的所有字串并返回一个列表
        for new_url in new_urls:    #从new_urls列表中提取元素
            if new_url not in storage:    #如果该元素不在集合中
                queue.extend(new_urls)    #则将其加入对列中
        count = count - 1    #计数器减一
    except Exception as e:    #如果try部分出错了，就执行except部分
        print(e)
    else:    #如果try部分正确就执行else部分
        print(url)
```

* 根据去重的广度遍历的结果显示，不去重广度遍历和深度遍历爬取10网页只显示六个是因为重复爬取而形成闭环所致。
* 由此可见，爬取时去重很重要

这里对结果进行分析。

### (选)将抓取网页的去重深度遍历算法封装成对象（类），并测试。 

```python
# 这里编写代码
class Spider():
    def __init__(self,url):
        self.url = url
    def dgf(self):
        import requests,re    #导入requests库，re库（正则表达式）
        count = 10    #记数器设置为10
        r = re.compile(r'href=[\''']?(http[^\'''>]+)')    #调用re库中的compile函数，其将正则表达式创建为模式对象
        headers = {
              'User-Agent':'Mozilla/4.0(compatible; MSIE 5.5; Windows NT)', 'hose':'httpbin.org'
         }       #伪装成火狐浏览器
        seed = self.url     #设置种子网站
        queue = [seed]         #将种子网站放入一个队列中
        storage = set()    #构造一个集合对象
        while len(queue) > 0 and count > 0:    #这是一个while循环。当队列为空时且记数器为0时结束
            try:
                url = queue.pop(-1)    #使用pop方法弹出队列的最后一个元素
                html = requests.get(url,params=headers).text    #使用requests库中的get函数发送请求，并将返回的请求结果字符串化
                storage.add(url)   #将爬取的元素加入到集合storage中
                new_urls = r.findall(html)    #使用re库中的findall函数，将在字符串中找到的正则表达式所匹配的所有字串并返回一个列表
                for new_url in new_urls:
                    if new_url not in storage:
                        queue.extend(new_urls)    #将新返回的列表加入到原始列表中
                count = count - 1    #计数器减一
            except Exception as e:    #如果try部分出错了，就执行except部分
                print(e)
            else:    #如果try部分正确就执行else部分
                print(url)
```

```python
# 测试
url = Spider('http://www.baidu.com/')
url.dgf()
```
