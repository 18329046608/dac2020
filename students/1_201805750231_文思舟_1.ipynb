### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。

```python
import urllib.request as ur
import re

count = 0
r = re.compile(r'href=[\'"]?(/item[^\'" >]+)')
# 抽取所需链接信息的正则语言规则
seed = "/item/%E7%BD%91%E7%BB%9C%E7%88%AC%E8%99%AB" 
# 数据采集词条
stack = [seed]
# 设置种子链接的栈（使用列表模拟栈）
storage = {}
while count < 5:
    try:
        url = stack.pop(-1) 
        # 取出栈的最后一条URL
        html = ur.urlopen("https://baike.baidu.com"+url).read().decode('utf-8') 
        # 对URL进行拼接
        new_urls = r.findall(html) 
        # 提取当前网页下的所有链接URL信息
        print(new_urls)
        stack.extend(new_urls) 
        # 将新提取的链接信息入队列
        storage[url] = len(new_urls)
        count += 1
    except Exception as e:
        print(url)
        print(e)
```

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
#不去重
import urllib.request as ur
import re

count = 0
r = re.compile(r'href=[\'"]?(/item[^\'" >]+)')
# 抽取所需链接信息的正则语言规则
seed = "/item/%E6%95%B0%E6%8D%AE%E9%87%87%E9%9B%86"
# 数据采集词条
queue = [seed]
# 设置种子链接的队列（使用列表模拟队列）
storage = {}
while count < 5:
    try:
        url = queue.pop(0)
        # 取出队列最后一条URL
        html = ur.urlopen("https://baike.baidu.com"+url).read().decode('utf-8')
        # 对URL进行拼接
        new_urls = r.findall(html)
        # 提取当前网页下的所有链接URL信息
        print(new_urls)
        queue.extend(new_urls)
        # 将新提取的链接信息入队列
        storage[url] = len(new_urls)
        count += 1
    except Exception as e:
        print(url)
        print(e)
```

```python
#去重
import urllib.request as ur
import re

count = 0
r = re.compile(r'href=[\'"]?(/item[^\'" >]+)')
# 抽取所需链接信息的正则语言规则
seed = "/item/%E6%95%B0%E6%8D%AE%E9%87%87%E9%9B%86"
# 数据采集词条
queue = [seed]
# 设置种子链接的队列（使用列表模拟队列）
used = set()
# 设置一个集合，保存已经抓取过的URL
storage = {}
while len(queue) > 0 and count < 5:
    try:
        url = queue.pop(0)
        html = ur.urlopen("https://baike.baidu.com"+url).read().decode('utf-8') 
        storage[url] = html
        used.add(url)
        # 将已经抓取过的添加到used集合中
        new_urls = r.findall(html)
        # 将新发现的未抓取过的添加到queue中
        for new_url in new_urls:
            if new_url not in used and new_url not in queue :
                queue.extend(new_url)
        count += 1
    except Exception as e:
        print(url)
        print(e)
```
