### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
##不去重
import requests,re
count=10 
r=re.compile(r'href=[\'"]?(http[^\'" >]+)')
seed='https://www.baidu.com/'  ##种子网站
queue=[seed]
storage={}
while len(queue)>0 and count >0:
            try:
                url=queue.pop(0)
                html=requests.get(url).text
                new_urls=r.findall(html)
                for new_url  in new_urls:  ##实现广度遍历
                    if new_url not in queue:
                        print(new_url)  ##输出网址
                storage[url]=html 
                queue.extend(new_urls)
                count-=1
            except Exception as e:
                print(url)
```

```python
##去重
import requests,re
count=3
allcount=1
recount=0
r=re.compile(r'href=[\'"]?(http[^\'" >]+)')
seed='https://www.baidu.com/'  ##种子网站
queue=[seed]
storage={}
used=set()
while len(queue)>0 and count >0:
            try:
                url=queue.pop(0)
                html=requests.get(url).text 
                storage[url]=html 
                used.add(url)
                new_urls=r.findall(html)
                for new_url in new_urls:
                    allcount+=1
                    if new_url not in used and new_url not in queue:   ##去重
                        queue.append(new_url)  ##输出去重后的网址链接
                        print(new_url)
                    else:
                        print('重复的:'+new_url) ##输出重复的网址链接
                        recount+=1
                count-=1
            except Exception as e:
                print(url)
                #print(e)
print('重复网站:'+str(recount))
print('总网站：'+str(allcount))
```
