### 问题1：用``urllib.requests``库发起一次``get``请求，输出``response``文本信息。

```python
#法一
import urllib.request#导入urllib.request库
url_get='https://www.runoob.com/regexp/regexp-tutorial.html'
response = urllib.request.urlopen(url_get)# 向指定的url发送请求，并返回服务器响应的类文件对象
html = response.read()# 服务器返回的类文件对象支持Python文件对象的操作方法，如read()方法读取文件全部内容，返回字符串
print(html.decode('utf-8'))# 打印响应的内容
```

```python
#法二
import urllib.request
request = urllib.request.Request("http://admin.bxcker.com")# url 作为Request()方法的参数，构造并返回一个Request对象
response = urllib.request.urlopen(request)# Request对象作为urlopen()方法的参数，发送给服务器并接收响应
html = response.read()
print(html.decode('utf-8'))#.decode('utf-8')选择文字编码方式
```

### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
import requests#导入requests库
url='https://www.runoob.com/regexp/regexp-tutorial.html'#要爬取的网页
payload_tuples=[('key1','value1'),('key1','value2')]
#payload_tuples={'key1':('value1','value2')}
send_headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:23.0) Gecko/20100101 Firefox/23.0'}#定制报文头，修改User-Agent，伪装成火狐浏览器
r1=requests.post(url,headers=send_headers,data=payload_tuples)
print(r1.text)
```

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。

```python
import requests#导入requests库
import re#导入re库
count=10
r=re.compile(r'href=[\'"]?(http[^\'">]+)')#编译正则表达式
seed='https://httpbin.org/'#设置种子网站
stack=[seed]#种子网站入栈
storage={}
while len(stack)>0 and count>0:
    try:
        url=stack.pop(-1)
        html=requests.get(url).text
        new_urls=r.findall(html)
        stack.extend(new_urls)
        storage[url]=len(new_urls)
        count-=1
    except Expection as e:
        print(url)
        print(e)
for i in stack:
    print(i)
```

```python
from collections import Counter
url_count=Counter(stack)
for url,count in url_count.most_common(10):
    print(url,count)
```

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
#去重广度遍历算法
import requests
import re
count = 10
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')#编译正则表达式
seed = 'https://cn.bing.com/' #
queue = [seed]
used = set() # 设置一个集合，保存已经抓取过的 URL
storage = {}
while len(queue) > 0 and count > 0:#hash去重
    try:
        url = queue.pop(0)
        html = requests.get(url).text
        storage[url] = html
        #将已经抓取过的URL存入used集合中
        used.add(url)
        new_urls = r.findall(html)
        #将新发新未抓取的URL添加到queue中
        for new_url in new_urls:
            if new_url not in used and new_url not in queue:
                queue.append(new_url)
        count=count-1
    except Exception as e:
        print(url)
        print(e)
for i in queue:
    print(i)
```

```python
from collections import Counter
url_count=Counter(queue)
for url,count in url_count.most_common(10):
    print(url,count)
```
