### 问题1：用``urllib.request``库发起一次``get``请求，输出``response``文本信息。

```python
#调用urllib库
import urllib 
#从urllib里调用request模块
from urllib import request
#指定爬取网页
url="http://news.baidu.com//get"
#输入参数
headers={"User-Agent":'Mozilla/4.0(compatible;MSIE 5.5;Window 10.0)',"host":'www.baidu.com'}
dict={"name":'百度新闻'}
data=bytes(urllib.parse.urlencode(dict),encoding='utf-8')
#调用request模块下的Request函数爬取网页信息
req=request.Request(url=url,data=data,headers=headers,method='GET')
#将爬取到的数据返回给response
response=request.urlopen(req)
#将爬取到的数据以文本形式输出
print(response.read().decode("utf-8"))
```

### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
#导入urllib,request,parse
import urllib 
from urllib import request,parse
#将网页地址加上爬取方式post返回给url
url="http://httpbin.org/post"
#输入参数
headers={"User-Agent":'Mozilla/4.0(compatible;MSIE 5.5;Window 10.0)',"host":'httpbin.org'}
dict={"name":'新型冠状病毒最新消息'}
data=bytes(urllib.parse.urlencode(dict),encoding='utf-8')
#用Request库伪装火狐浏览器发起一次post请求
req=request.Request(url=url,data=data,headers=headers,method='POST')
#将爬取到的数据返回给response
response=request.urlopen(req)
#以文本形式输出response
print(response.read().decode("utf-8"))
```

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
#导入requests，re库
import requests,re
#初始化深度
count = 3
#用正则表达式达到pattern对象
r = re.compile(r'href=[\'"]?(http[^\'">]+)')
#设定一个种子网站
seed = 'https://baidu.com/more/'
#将种子网站放入队列
queue = [seed]
#定义一个集合来放爬取到的url
storage = {}
while len(queue)>0 and count >0:
    try:
        #弹出列表第一个url
        url = queue.pop(0)
        #将爬取到的url以text形式保存到html
        html = requests.get(url).text
        #将html中的数据保存到storage[]中
        storage[url] = html
        #继续向下搜索
        urls = r.findall(html)
        #输出url个数
        print(url + "下的url的数量为：" + str(len(urls)))
        count -= 1
    #捕获异常
    except Exception as e:
        print(e)
for i in range(len(urls)):
    print(urls[i])
```

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
#去重
#导入urllib中的request模块并记为ur
import requests 
#导入re库
import re
#控制层数
count = 3
#初始化数据存储结构
r = re.compile(r'href=[\'"]?(http[^\'">]+)')
#给定种子网站
seed = 'https://httpbin.org/'
#将种子网站放入队列
queue = [seed]
storage = {}
used=set()
#判断是否进行爬取
while len(queue)>0 and count!=0:
    try:
        url = queue.pop(0)
        #用urlopen爬取到的数据以utf-8形式保存到html
        used.add(url)
        html = requests.get(url).text
        #把html放入已定义好的列表中
        storage[url] = html
        #继续爬取
        new_urls = r.findall(html)
        #目前爬取到的层数小于给定层数，则执行以下语句
        for new_url in new_urls:
            #判断此url是否已被爬取过
            if new_url not in used and new_url not in queue:
                #如果没有则加入队列
                queue.append(new_url)
        #计数加一
        count -= 1
    #捕获异常
    except Exception as e:
        print(e)
        print(url)
used=list(used)
used.extend(queue) #由于是广度优先所以只要按照stack的顺序输出
for i in range(len(used)):
    print(used[i])#输出url
print(len(used))
```
