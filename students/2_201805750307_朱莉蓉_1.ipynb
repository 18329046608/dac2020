### 问题1：用``urllib.requests``库发起一次``get``请求，输出``response``文本信息。

```python
# 这里编写代码
import urllib.request  #导入urllib.request库
url = 'https://www.cnki.net/' #选择一个网址
r = urllib.request.Request(url=url,method='GET') #构造Request实例，以GET方式请求，GET对大小写敏感
response = urllib.request.urlopen(r) #发送请求
print(response.read().decode('utf-8'))  #打印结果
```

### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
# 这里编写代码
import requests
t1=[('k1','v1'),('k2','v2')]
headers = {
    'Use-Agent':'Mozilla/4.0(compatible; MSIE 5.5; Windows NT)', 'hose':'httpbin.org'
}#伪装成火狐浏览器
response = requests.post('http://www.cninfo.com.cn/new/index',data=t1,headers=headers)#发送post请求
print(response.text)#打印结果
```

伪装成火狐浏览器，是躲避反爬虫策略最简单的方式

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
```

```python
import requests,re  #导入requests库和re库
count = 10   #计数器
r = re.compile(r'href=[\''']?(http[^\'''>]+)')  #调用re库中的compile函数，抽取所需链接信息的正则语言
seed ='http://www.baidu.com/'  #自选的一个种子网站
stack = [seed]   #设置一个栈，放未爬取的网站
storage = {}
while len(stack)>0 and count>0:
    try:
        url = stack.pop(-1)    #拿出栈顶的元素
        html = requests.get(url).text
        new_url = r.findall(html)#使用re库中的findall函数，找到所有满足r的URL
        stack.extend(new_url)   #注意此处用extend
        storage[url] = len(new_url)
        count-=1   #计数器减1
    except Exception as e:  #try部分错误，则执行except部分
        print(url)
        print(e)
```

深度遍历使用数据结构中的栈结构，优先拿取栈顶的元素（即FILO）  
注意到深度遍历和广度遍历出的网站顺序是不一样的，深度遍历收集的网站越来越远，会收集很多不相关的信息

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
```

```python
#去重广度遍历算法
import requests,re
count = 10
r = re.compile(r'href=[\''']?(http[^\'''>]+)')
seed = 'http://www.baidu.com/' #种子网站
used = set()#设置一个集合装已经爬取过的网页
queue = [seed] #将种子网站加入队列中
while len(queue)>0 and count>0:
    try:
        url = queue.pop(0)   #用pop取出第一个进入队列的元素
        html = requests.get(url).text
        used.add(url) #已经遍历过，加入used中
        new_urls = r.findall(html) #使用findall函数，将在字符串中找到的正则表达式所匹配的所有字串并返回
        for new_url in new_urls:
            if new_url not in used and new_url not in queue:#！去重的重要步骤！不在已经爬取的集合中，也不在队列中
                queue.append(new_url)   #则添加在队列中
        count-=1
    except Exception as e:  #try部分错误，则执行except部分
        print(url)
        print(e)
```

```python
#不去重的广度遍历算法，和去重主要差在上述重要步骤，注释略
import requests,re
r = re.compile(r'href=[\''']?(http[^\'''>]+)')
count = 10
seed = 'http://www.baidu.com/'
queue = [seed]
while len(queue)>0 and count>0:
    try:
        url = queue.pop(0)
        html = requests.get(url).text
        new_url = r.findall(html)
        queue.extend(new_url)
        count-=1
    except Exception as e: 
        print(url)
        print(e)
```
