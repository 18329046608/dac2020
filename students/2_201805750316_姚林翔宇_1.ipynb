### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
# 去重BFS
def BFS1(seed):
    count = 5
    r = re.compile(r'href=[\'"]?(http[^\'" >]+)')
    used = set()  # 重复的放入集合
    queue = [seed]
    total_urls = []  # 总url数
    while len(queue) > 0 and count > 0:
        try:
            url = queue.pop(0)
            used.add(url)
            total_urls.append(url)
            html = requests.get(url).text
            new_urls = r.findall(html)
            for new_url in new_urls:  # 去重
                if new_url not in used and new_url not in queue:
                    queue.append(new_url)
                    total_urls.append(new_url)
            count -= 1
        except Exception as e:
            print(url)
            print(e)
    print('去重后总url数为：',end='')
    print(len(total_urls))
    print(total_urls)
    
# 不含去重BFS
def BFS2(seed):
    count = 5
    r = re.compile(r'href=[\'"]?(http[^\'" >]+)')
    queue = [seed]
    total_urls = []
    while len(queue) > 0 and count > 0:
        try:
            url = queue.pop(0)
            html = requests.get(url).text
            new_urls = r.findall(html)
            queue.extend(new_urls)
            total_urls.extend(new_urls)
            count -= 1
        except Exception as e:
            print(url)
            print(e)
    print('去重前总url数为：', end='')
    print(len(total_urls))
    print(total_urls)
    

BFS2('http://httpbin.org/')
```
