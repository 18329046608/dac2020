### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
#使用urllib.request
from urllib import request,parse
url = "http://httpbin.org/post"
headers = {"User-Agent":'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)',"host":'httpbin.org'}  #伪装成火狐浏览器
dict = {"name":"Hackdata"}
data = bytes(parse.urlencode(dict),encoding="utf8")
req = request.Request(url=url,data=data,headers=headers,method="POST") #发起post请求
response = request.urlopen(req)
print(response.read().decode("utf-8"))  #read方法得到的响应的数据是二进制数据，使用decode解码成utf-8编码
```

### 用``Requests``库中的``session``对象发出``get``请求，设置``cookies``并获取，输出获取的``cookies``内容。 

```python
import requests
# 创建一个session对象 
s = requests.Session() 
# 用session对象发出get请求，设置cookies 
s.get('http://httpbin.org/cookies/set/sessioncookie/123456789')    #cookie保留在s中
# 用session对象发出另外一个get请求，获取cookies 
r = s.get("http://httpbin.org/cookies")  #再次访问时会保留cookie
# 显示结果 
print(r.text)
```

结果分析：Session可以使用get，post等方法，Session对象在请求时允许你保留一定的参数和自动设置cookie

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
import requests, re
count = 10    #层数
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')  # 抽取所需链接信息的正则表达式
seed = 'https://baike.baidu.com' # 爬取百度百科主页
stack = [seed]  # 设置种子链接的栈（使用列表模拟栈）
storage = {}
while len(stack) > 0 and count > 0:
    try:
        url = stack.pop(-1)   # 取出栈的最后一条URL
        html = requests.get(url).text
        new_urls = r.findall(html) # 提取当前网页下的所有链接URL信息
        stack.extend(new_urls) # 将新提取的链接信息入栈
        storage[url] = len(new_urls)  
        count -= 1
    except Exception as e:
        print(url)
        print(e)
#统计URL管理器中各URL的数量
from collections import Counter
url_count = Counter(stack)
for url,count in url_count.most_common(10):
    print(url,count)
```

结果分析：
使用list模拟栈，实现DFS算法

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
#不去重（广度遍历）
import requests, re
count = 10   #层数
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')  # 抽取所需链接信息的正则表达式
seed = 'https://baike.baidu.com' # 爬取百度百科主页
queue = [seed]  # 设置种子链接的队列（使用列表模拟队列）
storage = {}
while len(queue) > 0 and count > 0:
    try:
        url = queue.pop(0)   # 取出队列最后一条URL
        html = requests.get(url).text
        new_urls = r.findall(html)  # 提取当前网页下的所有链接URL信息
        queue.extend(new_urls)   # 将新提取的链接信息入队列
        storage[url] = len(new_urls)
        count -= 1
    except Exception as e:
        print(url)
        print(e)
#统计URL管理器中各URL的数量
from collections import Counter
url_count = Counter(queue)
for url,count in url_count.most_common(10):
    print(url,count)
```

结果分析：上面的代码用list模拟了queue，用len方法判断队列是否为空，每次发现新的URL都添加到queue的末尾（extend），每次都抓取队首的URL

```python
#去重（广度遍历）
import requests, re
count = 10     #层数
r = re.compile(r'href=[\'"]?(http[^\'" >]+)')  # 抽取所需链接信息的正则表达式
seed = 'https://baike.baidu.com' # 爬取百度百科主页
queue = [seed]   # 设置种子链接的队列（使用列表模拟队列）
used=set() #设置一个集合，保存已经抓取过的URL
storage = {}
while len(queue) > 0 and count > 0:
    try:
        url = queue.pop(0)   # 取出队列最后一条URL
        html = requests.get(url,timeout=5).text
        storage[url]=html
        used.add(url)  #将已经抓取过的添加到used 集合中
        new_urls = r.findall(html)  # 提取当前网页下的所有链接URL信息
        #将新发现的未抓取过的URL添加到queue中
        for new_url in new_urls:
            if new_url not in used and new_url not in queue:
                queue.append(new_url)
        count-=1
    except Exception as e:
        print(url)
        print(e)
#统计URL管理器中各URL的数量
from collections import Counter
url_count = Counter(queue)
for url,count in url_count.most_common(20):
     print(url,count)
```

结果分析：使用set()方法进行URL去重和层数控制

将爬取过的URL存放在一个元素集合中，在进行新的爬取之前将目标URL与爬取过的集合进行对比，只爬取元素集合中没有的URL，就可以完成去重处理。
并且设置限制爬取链接的层数。
