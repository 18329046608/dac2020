### 问题1：用``urllib.requests``库发起一次``get``请求，输出``response``文本信息。

```python
# 这里编写代码
# 这里编写代码
import urllib.request   #导入urllib.request库
url="http://data.stats.gov.cn/easyquery.htm?cn=E0103"  #输入要爬取的网页国家数据2018年全国各省地区生产总值统计
response=urllib.request.urlopen(url=url)  #使用urlopen函数向网页发出请求,返回响应值
print(response.read().decode('utf-8'))  #使用read函数读取相应内容，将响应内容转化为utf-8格式
```

### 用``Requests``库伪装成火狐浏览器发起一次``post``请求，输出``response``文本信息。

```python
# 这里编写代码
import urllib.request
url="http://data.stats.gov.cn/easyquery.htm?cn=E0103"
headers={   'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) '
                         'Chrome/51.0.2704.63 Safari/537.36'} #伪装成火狐浏览器，给请求头添加头部信息
req=urllib.request.Request(url=url,headers=headers)  #增加了头部信息，用Request类构建新的一个完整请求
response=urllib.request.urlopen(req)
print(response.read().decode('utf-8'))
```

这里对结果进行分析：这里对结果进行分析：urlopen只能完成一些简单基本的请求，如果需要更加完整的请求则需要用Request,此时urlopen的参数不再是URL而是一个Request类,伪装身份时需要重新增加头部信息

### 用``Requests``库中的``session``对象发出``get``请求，设置``cookies``并获取，输出获取的``cookies``内容。 

```python
# 这里编写代码
import requests
url="http://data.stats.gov.cn/easyquery.htm?cn=E0103/cookies"
s=requests.Session() #创建一个Session对象
r=s.get(url=url)  #Session对象发出get请求
print(r.cookies)  #打印RequestCookieJar类型的cookies
print(r.text)  #打印出cookies的内容
```

这里对结果进行分析：Session对象发出get请求后得到的是CookieJar类型的对象，若要输出文本形式组需要进一步转换

### 实现抓取网页的不去重深度遍历算法，自选合适的种子网站和相关参数，输出结果。


```python
# 这里编写代码
import requests,re
def fun1(url,deep):
    r=re.compile(r'href=[\'"]?(http[^\'">]+)') #给定正则化的格式
    count=0
    queue=[url]
    while count<deep and len(queue)>0:
        url=queue.pop(0)  #队头出队，进行深度遍历
        html=requests.get(url).text  #得到该url的文本信息
        new_urls=r.findall(html)  #正则表达式re findall方法以列表的形式返回能匹配的子串
        print(url)
        for new_url in new_urls:  #对第一层中的元素进行深度递归
            queue.append(new_url) #将第一层中的url入队
            fun1(new_url,deep-1)
        count=count+1
    return
seed="https://cn.bing.com/"  #给定抓取的网页
fun1(seed,3)
```

这里对结果进行分析：深度遍历需要对队头元素实现递归算法，给定深度后，在深度范围内进行递归，re.findall返回的是与在re.comile中匹配的字符串形成的列表形式结果

### 编写抓取网页的广度遍历算法（含去重和不去重），自选合适的种子网站和相关参数，输出结果。 

```python
# 这里编写代码
#不去重
import requests,re
r=re.compile(r'href=[\'"]?(http[^\'">]+)') #给定正则化的格式
seed="https://cn.bing.com/"  #给定抓取的网页
queue=[seed]
count=3
while count>0 and len(queue)>0:
    url=queue.pop(0)  #队头出队
    html=requests.get(url).text  #得到该url的文本信息
    new_urls=r.findall(html)  #正则表达式re findall方法以列表的形式返回能匹配的子串
    for new_url in new_urls:  #对第一层中的元素进行打印
        print(new_url)
        queue.append(new_url) #将第一层中的url入队
    count=count-1
```

```python
#去重
import requests,re
r=re.compile(r'href=[\'"]?(http[^\'">]+)') #给定正则化的格式
seed="https://cn.bing.com/"  #给定抓取的网页
queue=[seed]
count=3
used=set()  #用于存储已经爬取过的url
storage={}  #暂时存储某一层下的url
while count>0 and len(queue)>0:
    url=queue.pop(0)  #队头出队，进行深度遍历
    html=requests.get(url).text  #得到该url的文本信息
    storage[url]=html
    used.add(url)
    new_urls=r.findall(html)  #正则表达式re findall方法以列表的形式返回能匹配的子串
    for new_url in new_urls:  #对第一层中的元素进行深度递归
        if new_url not in used and new_url not in queue:
            print(new_url)
            queue.append(new_url)
    count=count-1
```

这里对结果进行分析:去重需要指定一个放某一层抓取的所有网页的集合,再指定一个放没有被抓取过的url的队列用于去重
